```python
class Counter : 
    # 요구사항 1번 함수
    def  __init__(self, count = 0, step = 1) :
        self.count = count
        self.step = step
        
    # 요구사항 2번 함수    
    def __str__(self) :
        return "[Count(step : %d)] %d" %(self.step, self.count)
    
    # 요구사항 3번 함수
    def incr(self) :
        self.count = self.count + self.step
        
    # 요구사항 4번 함수
    def __call__(self) :
        self.incr()
        
    # 요구사항 5번 함수
    def __add__(self, num) :
        return Counter(self.count + num, self.step)
    
    def __sub__(self, num) :
        return Counter(self.count - num, self.step)
    
    # 요구사항 6번 함수
    def __lt__(self, other) :
        return self.count < other
    def __gt__(self, other) :
        return self.count > other
    def __eq__(self, other) :
        return self.count == other
    
# 요구사항 1번
print("---요구사항 1번---")
c = Counter(10)
d = Counter(10, 2)

# 요구사항 2번
print("---요구사항 2번---")
print(c)
print(d)

# 요구사항 3번
print("---요구사항 3번---")
c.incr()
d.incr()
print(c)
print(d)

# 요구사항 4번
print("---요구사항 4번---")
c()
d()
print(c)
print(d)

# 요구사항 5번
print("---요구사항 5번---")
c = c + 5
d = d - 5
print(c)
print(d)

# 요구사항 6번
print("---요구사항 6번---")
print(c > 10) 
print(d > 10)
print(c < 10)
print(d < 10)
print(c == 17)
print(d != 9)
```

#### 4번 문제

다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

#### 4번 설명
먼저 MySet 클래스의 생성자 __init__()은 객체가 생성됨과 동시에 자동으로 호출이 됩니다. 이 때 MySet 클래스의 생성자는 리스트를 인자로 전달받는데, 반복문과 append() 메서드를 통해서 list의 요소들 하나씩 저장합니다.
그리고 eliminate_duplicate() 메소드를 호출하여 중복을 제거합니다. 중복을 제거하는 과정은 위 코드를 보시다시피 간단합니다. eliminate_duplicate() 메소드는 반복문과 조건문을 통해서 중복 값인지 검사를 하고나서 중복값이 아닐 경우에만 리스트에 삽입합니다. 그리고 나서 원래 list변수를 빈 리스트로 초기화 한 후 중복값을 제외한 요소들을 다시 삽입합니다.
마지막으로 __str__()은 객체의 내용을 문자열 형식으로 리턴하는데, 반복문을 통해서 각 요소들을 문자열 형태로 저장합니다 그리고 모든 문자열 사이에는 쉼표도 삽입합니다. 위 반복문이 끝나고 슬라이싱을 이용하여 마지막 요소 이후에 추가된 띄어쓰기와 쉼표( ,)를 제외한 후 result를 다시 초기화하고 리턴합니다. end값이 len()메소드에서 2를 빼준 값인 이유는 띄어쓰기와 쉼표는 크기가 각각 1로, 총 2만큼의 길이를 지워야 하기 때문입니다. 이렇게 필요없는 문자열들을 지운 뒤 마지막에 "}" 문자열을 추가해준 result를 리턴해주면 __str__()도 완성입니다!
그리고 @staticmethod는 클래스에서 직접 접근할 수 있는 메소드를 의미합니다.
마지막으로 if__name__ == "__main__"는 메인 함수의 선언, 시작을 의미하는 코드로 인터프리터에서 직접 실행했을 경우에만 실행할 것을 의미합니다.

#### 5번 문제

4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오.

정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.

요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.

> u = s | t

> print(u)

> MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9}

요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.

> u = s & t

> print(u)

> MySet: {2, 3}

요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.

> s = MySet([1, 2, 3])

> t = MySet([3, 4, 5])

> u = s - t

> print(u)

> MySet: {1, 2}

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
    
    # 요구사항 1
    # 요구사항 1은 __or__()를 중복 정의했습니다. 
    # 합집합을 반환하는 요구사항으로 MySet(self+other)를 리턴합니다.
    def __or__(self, other):
        return MySet(self+other)
        
    # 요구사항 2
    # 요구사항 2는 __ and__()를 중복 정의했습니다. 
    # 먼저 self 객체를 s라는 변수에 리스트 형태로 저장합니다. 
    # 그리고 반복문과 if문을 통해서 겹치지 않는 요소가 존재한다면 remove()메서드로 지운 뒤 s객체를 MySet객체로 리턴합니다.
    def __and__(self, other):
        s = list(self)
        for e in self:
            if e not in other:
                s.remove(e)
        return MySet(s)
    
    # 요구사항 3
    # 요구사항 3은 __sub__()를 중복 정의했습니다. 
    # 위 요구사항 2와 굉장히 유사항 형태로 먼저 self 객체를 s라는 변수에 리스트 형태로 저장합니다. 
    # 그리고 반복문과 if문을 토앻서 겹치는 요소가 존재하면 remove()메서드로 지운 뒤 s 객체를 MySet객체로 리턴합니다.
    def __sub__(self, other):
        s = list(self)
        for e in self:
            if e in other:
                s.remove(e)
        return MySet(s)
        
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    
    # 요구사항 1
    u = s|t
    print(u)
    print()
    
    # 요구사항 2
    u = s&t
    print(u)
    print()
    
    # 요구사항 3
    s = MySet([1, 2, 3])
    t = MySet([3, 4, 5])
    u = s - t
    print(u)
```
